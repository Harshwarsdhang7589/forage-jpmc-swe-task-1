```python
#Harshwardhan Gaikwad
#Import required libraries

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier

# we had imported pandas for (data handling),sklearn for (model handling),accuracy(performance checking
# logistic regression,SVC,Adaboostclassifier,XGBoost:our 4 models
```

```python
# STEP 2: Load and Understand the Data
df = pd.read_csv("breast-cancer.csv")
df.head()
# df.head() shows the first 5 rows to check columns and values.
```

```python
# STEP 3: Data Preprocessing

# Check for missing values
print(df.isnull().sum())

# Converts diagnosis (M/B) to numeric values (1/0):


# Encode categorical data (if any)
le = LabelEncoder()
df["diagnosis"] = le.fit_transform(df["diagnosis"])  # 'M' becomes 1, 'B' becomes 0

# Drop unnecessary columns if any (like ID column)
df = df.drop(columns=["id", "Unnamed: 32"], errors="ignore")
```

```python
# STEP 4: Feature and Target Split

#split data into x(input features) and y as:(output labels)
X = df.drop("diagnosis", axis=1)  # features
y = df["diagnosis"]               # target


# Scales input for better performane.
#splits into 80% training ,20% testing

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# STEP 5: Train and Test Models

# Case :01 : Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))

# Support Vector Machine(SVM)
svm = SVC()
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)

print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))

# AdaaBoost
ada = AdaBoostClassifier()
ada.fit(X_train, y_train)
y_pred_ada = ada.predict(X_test)

print("AdaBoost Accuracy:", accuracy_score(y_test, y_pred_ada))

# XGBoost
xgb = XGBClassifier(eval_metric='logloss')
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)

print("XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))
```

```python
import gradio as gr
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt

# Your model components should already be loaded: lr, svm, ada, xgb, scaler, le, X_test, y_test

def predict_cancer(
    radius_mean, texture_mean, perimeter_mean, area_mean, smoothness_mean, compactness_mean, concavity_mean,
    concave_points_mean, symmetry_mean, fractal_dimension_mean, radius_se, texture_se, perimeter_se, area_se,
    smoothness_se, compactness_se, concavity_se, concave_points_se, symmetry_se, fractal_dimension_se,
    radius_worst, texture_worst, perimeter_worst, area_worst, smoothness_worst, compactness_worst, concavity_worst,
    concave_points_worst, symmetry_worst, fractal_dimension_worst
):
    input_data = [[
        radius_mean, texture_mean, perimeter_mean, area_mean, smoothness_mean, compactness_mean, concavity_mean,
        concave_points_mean, symmetry_mean, fractal_dimension_mean, radius_se, texture_se, perimeter_se, area_se,
        smoothness_se, compactness_se, concavity_se, concave_points_se, symmetry_se, fractal_dimension_se,
        radius_worst, texture_worst, perimeter_worst, area_worst, smoothness_worst, compactness_worst, concavity_worst,
        concave_points_worst, symmetry_worst, fractal_dimension_worst
    ]]
    input_data_scaled = scaler.transform(input_data)

    lr_pred = lr.predict(input_data_scaled)[0]
    svm_pred = svm.predict(input_data_scaled)[0]
    ada_pred = ada.predict(input_data_scaled)[0]
    xgb_pred = xgb.predict(input_data_scaled)[0]

    # Map predictions to desired output
    predictions = [lr_pred, svm_pred, ada_pred, xgb_pred]
    decoded_predictions = [ "Benign" if pred == 0 else "Malignant" for pred in predictions]

    return decoded_predictions[0], decoded_predictions[1], decoded_predictions[2], decoded_predictions[3]


def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    return {'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1-Score': f1}


def evaluate_models():
    results = {
        'Logistic Regression': evaluate_model(lr, X_test, y_test),
        'SVM': evaluate_model(svm, X_test, y_test),
        'XGBoost': evaluate_model(xgb, X_test, y_test),
        'AdaBoost': evaluate_model(ada, X_test, y_test)
    }

    results_df = pd.DataFrame(results).T  # Transpose so models are rows
    results_df.index.name = 'Model'

    # Add a 'Model' column to the DataFrame
    results_df.reset_index(inplace=True)

    # Plot
    fig, ax = plt.subplots(figsize=(8, 4.5))
    results_df.plot(kind='bar', ax=ax)
    ax.set_title('Model Accuracy Comparison')
    ax.set_ylabel('Accuracy')
    ax.set_ylim(0.8, 1.0)  # Start y-axis at 0.8
    ax.set_xticklabels(results_df['Model'], rotation=30, ha='right')
    ax.grid(True, linestyle='--', alpha=0.5)
    ax.legend(loc='lower right')
    plt.tight_layout()

    return gr.Plot(fig), results_df
```

```python
# Helper function to create a slider
def create_slider(label, minimum=0.0, maximum=100.0, step=0.1):
    return gr.Slider(label=label, minimum=minimum, maximum=maximum, step=step)

# Input components
input_components = [
    create_slider("radius_mean", 0, 30),
    create_slider("texture_mean", 0, 40),
    create_slider("perimeter_mean", 0, 200),
    create_slider("area_mean", 0, 2500),
    create_slider("smoothness_mean", 0, 0.5),
    create_slider("compactness_mean", 0, 1),
    create_slider("concavity_mean", 0, 1),
    create_slider("concave_points_mean", 0, 0.5),
    create_slider("symmetry_mean", 0, 1),
    create_slider("fractal_dimension_mean", 0, 0.2),
    create_slider("radius_se", 0, 5),
    create_slider("texture_se", 0, 5),
    create_slider("perimeter_se", 0, 10),
    create_slider("area_se", 0, 200),
    create_slider("smoothness_se", 0, 0.1),
    create_slider("compactness_se", 0, 0.5),
    create_slider("concavity_se", 0, 0.5),
    create_slider("concave_points_se", 0, 0.2),
    create_slider("symmetry_se", 0, 0.3),
    create_slider("fractal_dimension_se", 0, 0.1),
    create_slider("radius_worst", 0, 40),
    create_slider("texture_worst", 0, 50),
    create_slider("perimeter_worst", 0, 250),
    create_slider("area_worst", 0, 3000),
    create_slider("smoothness_worst", 0, 0.5),
    create_slider("compactness_worst", 0, 1),
    create_slider("concavity_worst", 0, 1),
    create_slider("concave_points_worst", 0, 0.5),
    create_slider("symmetry_worst", 0, 1),
    create_slider("fractal_dimension_worst", 0, 0.3)
]

# Output textboxes for predictions
output_components = [
    gr.Textbox(label="Logistic Regression"),
    gr.Textbox(label="SVM"),
    gr.Textbox(label="AdaBoost"),
    gr.Textbox(label="XGBoost")
]

with gr.Blocks(
    title="Breast Cancer Prediction",
    css="""
        .gradio-container {
            background-image: url('https://png.pngtree.com/thumb_back/fw800/background/20190223/ourmid/pngtree-blue-genetic-dna-medical-background-windbackground-display-boardbackgroundpsdcolorful-image_77077.jpg');
            background-size: cover;
            background-position: center;
        }
    """
) as iface:
    gr.Markdown("## Breast Cancer Prediction")
    gr.Markdown("Enter the values and click submit to get predictions from multiple models.")

    with gr.Row():
        with gr.Column(scale=5):
            for i in range(0, len(input_components), 4):
                with gr.Row():
                    for j in range(4):
                        if i + j < len(input_components):
                            input_components[i + j].render()
            submit_button = gr.Button("Submit")

        with gr.Column(scale=4):
            with gr.Row():
                for comp in output_components:
                    comp.render()

            with gr.Row():
                evaluation_table = gr.Dataframe(label="Evaluation Table")

            with gr.Row():
                evaluation_plot = gr.Plot(label="Evaluation Plot")

    # Actions
    submit_button.click(fn=predict_cancer, inputs=input_components, outputs=output_components)
    iface.load(fn=evaluate_models, inputs=None, outputs=[evaluation_plot, evaluation_table])

iface.launch(debug=True, share=True)
```
